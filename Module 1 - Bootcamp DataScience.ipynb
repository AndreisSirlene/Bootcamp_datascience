### Challenge 2
Add a column to show the 5 regions( Norte, Nordeste, Sul, Sudeste and Centro-Oeste) for all states.


```python
import pandas as pd
cleandata["2007/Ago"]
columnsDataFrame = cleandata.columns
index=["Norte, Sul, Sudeste, Centro-Oeste, Nordeste"]
dfperRegion = pd.DataFrame(columns=columnsDataFrame)

def total_col_region (states, column):
  total = 0.00
  for unit in states:
    total = total + cleandata.loc[unit, column]
  return total

for column in cleandata.columns:
  dfperRegion.loc["Norte", column] = total_col_region(list(['11 Rondônia', '12 Acre', '13 Amazonas', '14 Roraima', '15 Pará', '16 Amapá', '17 Tocantins']), column)
  dfperRegion.loc["Sul", column] = total_col_region(list(['41 Paraná', '42 Santa Catarina', '43 Rio Grande do Sul']), column)
  dfperRegion.loc["Sudeste", column] = total_col_region(list(['31 Minas Gerais', '32 Espírito Santo', '33 Rio de Janeiro', '35 São Paulo']), column)
  dfperRegion.loc["Centro-Oeste", column] = total_col_region(list(['50 Mato Grosso do Sul', '51 Mato Grosso', '52 Goiás', '53 Distrito Federal']), column)
  dfperRegion.loc["Nordeste", column] = total_col_region(list(['21 Maranhão', '22 Piauí', '23 Ceará','24 Rio Grande do Norte', '25 Paraíba', '26 Pernambuco', '27 Alagoas', '28 Sergipe', '29 Bahia']), column)
    
dfperRegion.head()
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

ax = dfperRegion.T.plot(figsize=(10,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("R$ {x:,.2f}"))

plt.style.use("fast")
plt.title("Expenses in Hospitalar procedures by Region")
plt.legend(loc="best", ncol=3, bbox_to_anchor=(1.70, 1))
plt.xlabel('Year/Month')
plt.ylabel("Total expend in Real")
plt.show()
```

## Challenge
Calculate the proportion expend by population of your state and another one to compare it.
Consider the population provide by IBGE available in the below website:
https://pt.wikipedia.org/wiki/Lista_de_unidades_federativas_do_Brasil_por_popula%C3%A7%C3%A3o

```python
dict_population_by_state = {'35 São Paulo':46289333,
'31 Minas Gerais':21331292,
'33 Rio de Janeiro':17366189,
'29 Bahia':14930634,
'41 Paraná':11516840,
'43 Rio Grande do Sul':11422973,
'26 Pernambuco':9616621,
'23 Ceará':9187103,
'15 Pará':8690745,
'42 Santa Catarina':7279638,
'52 Goiás':7142070,
'21 Maranhão':7127460,
'13 Amazonas':4228381,
'32 Espírito Santo':4064052,
'25 Paraíba':4039277,
'24 Rio Grande do Norte':3534165,
'51 Mato Grosso':3526220,
'27 Alagoas':3351543,
'22 Piauí':3281480,
'53 Distrito Federal':3055149,
'50 Mato Grosso do Sul':2809394,
'28 Sergipe':2318822,
'11 Rondônia':1796460,
'17 Tocantins':1590248,
'12 Acre':894470,
'16 Amapá':861773,
'14 Roraima':63118}
```

```python
exp_per_person_state = [[cleandata["Total"].loc[estado] / dict_population_by_state[estado], dict_population_by_state[estado]] 
                        for estado in dict_population_by_state.keys()]
sc_exp_per_person = pd.DataFrame(data=exp_per_person_state, index=dict_population_by_state.keys(),
                                 columns=['Expend per person', 'Total Population'])
sc_exp_per_person.sort_values(by='Total Population', ascending=False)['Expend per person'].plot(kind='bar', grid=True, 
                                                                                                figsize=(15,5))
plt.title("Expenses per person in each state, based in admissions by SUS between 2008 to 2020")
plt.show()
```

The graphic show us that consider the total population of each state and their total exenditure, when analyse the total expend per person compare with the state of Santa Catarina, the state of Roraima expend almost 5000 per person while Santa Catarina only expend 1000.
Sao Paulo is far the large state in terms of population in Brazil and expend less than 1000, consider that Sao Paulo is at a central position when come to industrial development and acess to services, the graphic clear show us the reality. By knowing the geographic and economic situation of the country, is understandable that when comes to health situation most cities in Roraima does not even have basic hospitals and are dependable of countries such as Sao Paulo, Brasilia and Minas Gerais to more specialized procedures, so states like Roraima expend considerable in transport of patients to diferent states and even with the health profissionals that are more dificult to hire in the North region of Brazil. 

```python
def cost_hab(population_X, state_X, population_Y, state_Y): 
  cost_X = cleandata["Total"].loc[state_X] / population_X
  cost_Y = cleandata["Total"].loc[state_Y] / population_Y
  sc_exp_per_person = pd.DataFrame(data=[cost_X, cost_Y], index=[state_X, state_Y], columns=['Expend per person'])
  sc_exp_per_person.plot(kind='bar', grid=True, figsize=(15,5))
```
